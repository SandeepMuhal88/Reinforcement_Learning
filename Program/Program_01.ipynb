```python
# import gymnasium as gym

# Example environment, assuming cliffEnv is a gym.Env object
cliffEnv = gym.make('CliffWalking-v1')  # Corrected line
state, info = cliffEnv.reset()

done = False
while not done:
    action = cliffEnv.action_space.sample()  # Use action_space.sample() for random actions
    next_state, reward, done, truncated, info = cliffEnv.step(action)

    state = next_state
    
cliffEnv.close()
```
